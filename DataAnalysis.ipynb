```python
#Filtering String in Pandas Dataframe
#considering sample example
df = pd.DataFrame({"var1": ["AA_2", "B_1", "C_2", "A_2"]})
df
```

```python
#Select rows having values starting from letter 'A',str[0] means 1st letter
newdf=df[df.var1.str[0]=='A']
newdf
```

```python
#Filter rows having string length greater than 3
newdf=df[df.var1.str.len()>3]
newdf
```

```python
#Select string containing letters A or B
df[df.var1.str.contains('A|B')]
```

```python
#Handle space in column name while filtering, lers rename var1 to var 1 for study
df.rename(columns={'var1':'var 1'}, inplace = True)
df
```

```python
#using backticks to work with the spaces in column name
newdf = df.query("`var 1` == 'AA_2'")
newdf
```
